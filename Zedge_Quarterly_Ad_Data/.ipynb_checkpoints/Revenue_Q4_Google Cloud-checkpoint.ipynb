```python
import pandas as pd
from google.cloud import bigquery
from datetime import datetime
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
# Big Query Command to pull data
%load_ext google.cloud.bigquery
```

```python
%%bigquery rev_totals
SELECT date, app, business_type, clicks, platform, connection, ad_name, ad_format, impressions, ad_revenue, tier, country 
   FROM `wallpaperboard-6dd71.revenue.daily`
WHERE 
app = 'Zedge' AND 
platform = 'Android' AND
ad_revenue > 1.00 AND
((date >= '2017-10-01' AND date <= '2017-12-31') OR 
(date >= '2018-10-01' AND date <= '2018-12-31'))
```

```python
q4_revenue_data = rev_totals[['date','app','connection','clicks','ad_name','impressions','ad_revenue','tier','country']]
```

```python
# GET MONTH AND YEAR
q4_revenue_data.date = pd.to_datetime(q4_revenue_data.date)

#q4_revenue_data.info()
q4_revenue_data['year'] = q4_revenue_data.date.dt.year
q4_revenue_data['month'] = q4_revenue_data.date.dt.month

#Ensure the Tier is a String
q4_revenue_data.tier = q4_revenue_data.tier.astype(str)

#Define a Higher Level Partner Field while Connection is manipulated
q4_revenue_data['partner'] = q4_revenue_data['connection']

q4_revenue_data.head()
```

```python
#q4_revenue_data.reset_index(inplace=True)
#q4_revenue_data
q4_revenue_data.head()
#q4_revenue_data.set_index('connection', inplace=True)
q4_revenue_data.head()

q4_revenue_data.connection.value_counts()
```

```python
q4_revenue_data.set_index('connection', inplace=True)

q4_revenue_data.drop('Tap Research', inplace=True)
q4_revenue_data.drop('Pollfish', inplace=True)

# changing index cols with rename() 
q4_revenue_data.rename(index = {"Liftoff": "Amazon TAM", 
                    "Rubicon": "Amazon TAM",
                    "Smaato" : "Amazon TAM",
                    "OpenX"  : 'Amazon TAM',
                    "Millenial" :'Other',
                    "Nexage"  : 'Other',
                    "Baidu"   : 'Other',
                    "Adview"  : 'Other',
                    "unity3d.com" : 'Other',
                    "Conversant"  : 'Other',
                    "Mobfox"    : 'Other'},
          
                                  inplace = True) 

q4_revenue_data.head(10)
```

```python
q4_revenue_data.reset_index(inplace=True)
```

```python
q4_agg = q4_revenue_data.groupby(['connection','year']).sum()
del q4_agg['month']
q4_agg

```

```python
q4_agg.reset_index(inplace=True)
q4_agg
```

```python
q4_agg.sort_values('ad_revenue', ascending=False, inplace=True )
```

```python
# Separate the 2017 and 2018 to calculate SOW
q42017 = q4_agg[q4_agg['year'] == 2017]
q42018 = q4_agg[q4_agg['year'] == 2018]

imp2017 = q42017['impressions'].sum()
imp2018 = q42018['impressions'].sum()
rev2017 = q42017['ad_revenue'].sum()
rev2018 = q42018['ad_revenue'].sum()

q42017['sow_imp'] = q42017.impressions / imp2017
q42017['sow_rev'] = q42017.ad_revenue /  rev2017

q42018['sow_imp'] = q42018.impressions / imp2018
q42018['sow_rev'] = q42018.ad_revenue /  rev2018

q42018
```

```python
# Combine the 2017 and 2018 files again for Graphing

frame = [q42018,q42017]
agg = pd.concat(frame)
agg
```

```python
from matplotlib.ticker import FuncFormatter

# fig = matplotlib.pyplot.gcf()
# fig.set_size_inches(18.5, 10.5)
# fig.savefig('test2png.png', dpi=100)

sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")

sns.set(rc={"font.size":14,"axes.labelsize":14})
# Rotate the labels on x-axis

sns.set_style('white')
sns.set_palette(sns.color_palette("Paired"))


ax = sns.barplot(x ='connection', y='sow_rev', hue='year', data=agg)


plt.title("Q4 2018 vs Q4 2017 \n Partner Share of Wallet \n", fontsize=20)
plt.xlabel('\n Partner')
plt.ylabel('Share of Wallet \n' )
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y))) 
ax.grid(linestyle='-', linewidth='0.2', color='gray')

# We change the fontsize of minor ticks label 
plt.tick_params(axis='both', which='major', labelsize=14)
plt.tick_params(axis='both', which='minor', labelsize=14)

# 5. Place legend to the right
#plt.legend(bbox_to_anchor=(1, 1), loc=2, fontsize=20)
plt.legend(loc='upper right', fontsize=25)
plt.savefig('Q4_2018_2017_SOW-1', dpi=100)



# plt.figure(figsize=(15,8))
# ax = sns.barplot(x="Word", y="Frequency", data=boxdata)

```
